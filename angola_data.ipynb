```python
from matplotlib import pyplot as plt
plt.style.use('classic')
%matplotlib inline
import pandas as pd
import seaborn as sns
sns.set()
```

```python
df = pd.read_csv('data/angola.csv')
df1 = pd.read_csv('data/ao.csv')
```

```python
df.head()
```

```python
df1.head()
```

```python
#Angola's population from 1952 to 2007
#Year VS Population  
f, ax = plt.subplots(figsize=(20, 7))
ax = sns.barplot(data=df, x='year', y='pop')
ax.set(xlabel='Year', ylabel='Population')
plt.savefig("img/angola_year_pop.png")
plt.show()
```

```python
#Population By Province
# Vertical
# Gráfico com as 12 ciddades com maior número de habitantes em Angola
sns.set_style('whitegrid')
sns.set_context("poster", font_scale=0.8)
f, ax = plt.subplots(figsize=(30, 10))
ax = sns.barplot(data=df1.head(12), x='city', y='population')
ax.set(xlabel='City', ylabel='Population')
plt.savefig("img/angola_city_pop.png")
plt.show()
```

```python
#Population By Province
#As 12 cidades com maior número de habitantes
f, ax = plt.subplots(figsize=(30, 10))
ax = sns.barplot(data=df1.head(12), x='city', y='population', hue="city")
ax.set(xlabel='City', ylabel='Population')
plt.savefig("img/angola_city_pop_hue.png")
plt.show()
```

```python
# Horizontal
#Population By Province
f, ax = plt.subplots(figsize=(30, 20))
ax = sns.barplot(data=df1, x='population', y='city')
ax.set(xlabel='Population', ylabel='City')
plt.savefig("img/angola_pop_city.png")
plt.show()
```

```python
#Angola's Life Expectancy from 1952 to 2007
#Year VS Life Expectancy
f, ax = plt.subplots(figsize=(20, 7))
ax = sns.barplot(data=df, x='year', y='lifeExp')
ax.set(xlabel='Year', ylabel='Life Expectancy')
plt.savefig("img/angola_year_life.png")
plt.show()
```

```python
#Angola's gross domestic product by years.
f, ax = plt.subplots(figsize=(20, 7))
ax = sns.barplot(data=df, x='year', y='gdpPercap')
ax.set(xlabel='Year', ylabel='GDP Percapital')
plt.savefig("img/angola_year_gdp.png")
plt.show()
```
